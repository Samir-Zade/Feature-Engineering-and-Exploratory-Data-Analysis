## **DATA INTERPOLATION TECHNIQUES**

- it is process of estimating unknown values within a dataset based on the known values.

**Types of data interpolation**

- 1) linear interpolation
- 2) cubic interpolation
- 3) polynomial interpolation
- 4) nearest neighbour interpolation

## 1) Linear Interpolation

```python
#import the libraries
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
x = np.array([1, 2, 3, 4, 5])
y = np.array([2, 4, 6, 8, 10])
```

```python
plt.scatter(x, y)
```

```python
#interpolate the data using linear interpolations techniques.
x_new = np.linspace(1, 5, 10) #create a 10 x datapoints
y_interp = np.interp(x_new, x, y) #(datapoints, for support, for support)
print(y_interp)
```

```python
plt.scatter(x_new, y_interp)
```

```python
x = np.array([1, 2, 3, 4, 5])
y = np.array([4, 5, 2, 5, 9])
plt.scatter(x, y)
```

## 2) Cubic Interpolation using SCIPY

```python
x = np.array([1, 2, 3, 4, 5])
y = np.array([1, 8, 27, 64, 125])
```

```python
!pip install scipy
```

```python
from scipy.interpolate import interp1d
```

```python
#create a cubic interpolate function
f = interp1d(x, y, kind = "cubic")
```

```python
#interpolate the data
x_new = np.linspace(1, 5, 10)
y_interp = f(x_new)
print(y_interp)
```

```python
plt.scatter(x, y)
```

```python
plt.scatter(x_new, y_interp)
```

### 3) Polynomial Interpolation

```python
x = np.array([1, 2, 3, 4, 5])
y = np.array([1, 8, 27, 64, 125])
```

```python
#interpolate the data using polynomial interpolation
p = np.polyfit(x, y, 2) #fit a 2nd degree polynomial to the data
```

```python
x_new = np.linspace(1, 5, 10) #create a new x value
y_interp = np.polyval(p, x_new)
```

```python
plt.scatter(x, y)
```

```python
plt.scatter(x_new, y_interp)
```
